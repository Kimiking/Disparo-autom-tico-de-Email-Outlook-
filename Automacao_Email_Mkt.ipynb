```python
import pandas as pd
import win32com.client as win32

# === 1. Ler o arquivo Excel ===
arquivo_excel = "Automacao.xlsx"
df = pd.read_excel(arquivo_excel)

# Verifica se as colunas esperadas est√£o presentes
colunas_esperadas = ["Empresa", "Nome", "E-mail"]
if not all(coluna in df.columns for coluna in colunas_esperadas):
    raise ValueError(f"A planilha deve conter as colunas: {colunas_esperadas}")

# === 2. Conectar ao Outlook ===
outlook = win32.Dispatch("Outlook.Application")

# === 3. Criar e enviar os e-mails ===
for _, linha in df.iterrows():
    empresa = linha["Empresa"]
    nome_completo = linha["Nome"]
    primeiro_nome = str(nome_completo).split()[0]  # <-- pega s√≥ o primeiro nome
    email_destino = linha["E-mail"]

    assunto = f"Proposta da {empresa}"

    corpo_email = f"""Ol√° {primeiro_nome},

Somos a Sandro Breves Consultoria, desenvolvemos um modelo de Atendimento Premium focado no mercado de Real Estate; com o objetivo de antecipar dores, encurtar prazos e garantir previsibilidade em cada etapa do processo.

Quem lidera essa opera√ß√£o √© o Sandro, com mais de 25 anos de experi√™ncia no setor, ao lado de uma equipe s√™nior preparada pra lidar com demandas de alta complexidade.

Gostaria de compartilhar um material que apresenta melhor como estruturamos esse conceito. Posso te encaminhar? 

Atenciosamente,
Wellington Oliveira
"""

    # Cria o e-mail no Outlook
    mail = outlook.CreateItem(0)
    mail.To = email_destino
    mail.Subject = assunto
    mail.Body = corpo_email  # Corpo em texto simples

    try:
        mail.Send()
        print(f"‚úÖ E-mail enviado para {primeiro_nome} ({email_destino}) - {empresa}")
    except Exception as e:
        print(f"‚ùå Erro ao enviar para {primeiro_nome} ({email_destino}): {e}")

print("\nüöÄ Processo conclu√≠do com sucesso!")
```
